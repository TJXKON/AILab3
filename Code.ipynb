**Import**

```python
from __future__ import print_function
import os
data_path = ['.']

import numpy as np
import pandas as pd
```

**Open CSV file**

```python
filepath = os.sep.join(data_path + ['cases_malaysia.csv'])
data = pd.read_csv(filepath)
```

**Create the mask for selecting date range**

```python
start_date = '2022-01-01'
end_date = '2022-06-30'
mask = (data['date'] >= start_date) & (data['date'] <= end_date)
```

**Update the dataframe within date range**

```python
data=data.loc[mask]
data
```
